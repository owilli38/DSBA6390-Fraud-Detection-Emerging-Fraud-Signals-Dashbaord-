```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""
    # Convert to lowercase
    text = text.lower()
    # Remove special characters and numbers, keep only letters and spaces
    text = re.sub(r'[^a-z\s]', '', text)
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

print("Text cleaning function 'clean_text' defined.")
```

```python
import numpy as np
import google.generativeai as genai

# Define the get_gemini_embeddings function
def get_gemini_embeddings(text):
    try:
        # Ensure text is a list, as embed_content can take a list of strings
        if isinstance(text, str):
            text = [text]

        response = genai.embed_content(
            model="models/gemini-embedding-001",
            content=text,
            task_type="retrieval_document" # Or "retrieval_query" depending on use case
        )
        # For a single text input, return the first embedding
        return response['embedding'][0]
    except Exception as e:
        print(f"Error generating embedding for text: {text}. Error: {e}")
        return None

fraud_concepts = [
    "financial fraud",
    "identity theft",
    "phishing scam",
    "money laundering",
    "cybercrime",
    "account takeover"
]

# Apply get_gemini_embeddings to each concept and convert to NumPy array
fraud_embeddings_list = [get_gemini_embeddings(concept) for concept in fraud_concepts]

# Filter out any None values if embedding generation failed for some concepts
fraud_embeddings_list = [emb for emb in fraud_embeddings_list if emb is not None]

# Only proceed if there are valid embeddings
if fraud_embeddings_list:
    fraud_embeddings_array = np.array(fraud_embeddings_list)

    # Calculate the mean across the rows (axis=0) to get a single fraud_embedding vector
    fraud_embedding = fraud_embeddings_array.mean(axis=0)

    print("Fraud embedding calculated successfully.")
else:
    print("No fraud embeddings could be generated.")
```

```python
import numpy as np
import ast # Required for safely evaluating string literals

# Function to safely parse an embedding value into a list of floats
def parse_embedding_to_list(embed_value):
    if isinstance(embed_value, str):
        try:
            # Use ast.literal_eval to safely convert string to Python literal (list, tuple, etc.)
            parsed = ast.literal_eval(embed_value)
            if isinstance(parsed, (list, tuple)):
                return [float(x) for x in parsed] # Ensure elements are floats
        except (ValueError, SyntaxError):
            return None # Not a valid string representation of a list
    elif isinstance(embed_value, np.ndarray):
        return embed_value.tolist()
    elif isinstance(embed_value, list):
        # Ensure it's a list of floats
        return [float(x) for x in embed_value]
    return None # For any other unexpected types or None values

# Apply the conversion to the 'embedding' column of the DataFrame
df['embedding'] = df['embedding'].apply(parse_embedding_to_list)

# Filter out any rows where the embedding could not be properly parsed or resulted in None
# This ensures that all elements in the list passed to np.array are valid lists of floats
valid_embeddings_list = [embed for embed in df['embedding'] if embed is not None and isinstance(embed, list)]

# Check for consistency in embedding dimensions before creating the NumPy array
if valid_embeddings_list:
    first_embedding_length = len(valid_embeddings_list[0])
    # Ensure all embeddings have the same dimension
    if not all(len(embed) == first_embedding_length for embed in valid_embeddings_list):
        print("Warning: Some embeddings have inconsistent lengths after cleaning. Proceeding with available consistent ones.")
        # Further filter to only include embeddings of the most common length, or raise an error.
        # For now, we will create the array from the cleaned list, which should be consistent if generation was.

    text_embedding_array = np.array(valid_embeddings_list)
    print(f"Shape of text_embedding_array: {text_embedding_array.shape}")
else:
    print("Warning: No valid embeddings found after cleaning. 'text_embedding_array' will be an empty array.")
    text_embedding_array = np.array([]) # Create an empty array if no valid embeddings
```

```python
max_length = 128  # Maximum length of input sentence to the model.
batch_size = 32
epochs = 2
```

```python
class BertInputBatchGenerator(tf_keras.utils.Sequence):

    def __init__(
        self,
        sentence_pairs,
        labels,
        batch_size=batch_size,
        shuffle=True,
        include_targets=True,
    ):
        self.sentence_pairs = sentence_pairs
        self.labels = labels
        self.shuffle = shuffle
        self.batch_size = batch_size
        self.include_targets = include_targets
        # Load our BERT Tokenizer to encode the text.
        # We will use bert-base-uncased pretrained model.
        self.tokenizer = transformers.BertTokenizer.from_pretrained(
            "bert-base-uncased", do_lower_case=True
        )
        self.indexes = np.arange(len(self.sentence_pairs))
        self.on_epoch_end()

    def __len__(self):
        # Return the number of batches per epoch.
        return len(self.sentence_pairs) // self.batch_size

    def __getitem__(self, idx):
        # Retrieves the batch of index.
        indexes = self.indexes[idx * self.batch_size : (idx + 1) * self.batch_size]
        sentence_pairs = self.sentence_pairs[indexes]

        # With BERT tokenizer's batch_encode_plus, batch of both the sentences are
        # encoded together and separated by [SEP] token.
        encoded = self.tokenizer.batch_encode_plus(
            sentence_pairs.tolist(),
            add_special_tokens=True,
            max_length=max_length,
            return_attention_mask=True,
            return_token_type_ids=True,
            padding='max_length', # Changed from pad_to_max_length=True
            truncation=True, # Added truncation=True
            return_tensors="tf",
        )

        # Convert batch of encoded features to numpy array.
        input_ids = np.array(encoded["input_ids"], dtype="int32")
        attention_masks = np.array(encoded["attention_mask"], dtype="int32")
        token_type_ids = np.array(encoded["token_type_ids"], dtype="int32")

        # Set to true if the data generator is used for training/validation.
        if self.include_targets:
            labels = np.array(self.labels[indexes], dtype="int32")
            return [input_ids, attention_masks, token_type_ids], labels
        else:
            return [input_ids, attention_masks, token_type_ids]

    def on_epoch_end(self):
        # Shuffle indexes after each epoch if shuffle is set to True.
        if self.shuffle:
            np.random.RandomState(42).shuffle(self.indexes)
```

```python
strategy = tf.distribute.MirroredStrategy()

with strategy.scope():
    # Encoded token ids from BERT tokenizer.
    input_ids = tf_keras.Input(
        shape=(max_length,), dtype=tf.int32, name="input_ids"
    )
    # Attention masks indicate which tokens are real and which are fake (padding).
    attention_masks = tf_keras.Input(
        shape=(max_length,), dtype=tf.int32, name="attention_masks"
    )
    # Token type ids indicate which tokens belong to Sentence 1 and which to Sentence 2.
    token_type_ids = tf_keras.Input(
        shape=(max_length,), dtype=tf.int32, name="token_type_ids"
    )
    # Loading pretrained BERT model.
    bert_model = transformers.TFBertModel.from_pretrained("bert-base-uncased")
    # Freeze the BERT model to reuse the pretrained features without modifying them.
    bert_model.trainable = False

    # Explicit conversion of KerasTensors to tf.Tensors
    bert_output = bert_model.bert(
        input_ids, attention_mask=attention_masks, token_type_ids=token_type_ids
    )
    sequence_output = bert_output.last_hidden_state
    pooled_output = bert_output.pooler_output

    # Add trainable layers on top of frozen layers to adapt the pretrained features on the new data.
    bi_lstm = tf_keras.layers.Bidirectional(
        tf_keras.layers.LSTM(64, return_sequences=True)
    )(sequence_output)
    # Applying hybrid pooling approach to bi_lstm sequence output.
    avg_pool = tf_keras.layers.GlobalAveragePooling1D()(bi_lstm)
    dropout = tf_keras.layers.Dropout(0.3)(avg_pool)
    output = tf_keras.layers.Dense(3, activation="softmax")(dropout)


    model = tf_keras.Model(
        inputs=[input_ids, attention_masks, token_type_ids], outputs=output
    )

    model.compile(
        optimizer=tf_keras.optimizers.Adam(),
        loss="categorical_crossentropy",
        metrics=["acc"],
    )


print(f"Strategy: {strategy}")
model.summary()
```

```python
import numpy as np

# Convert the list of 768-dim vectors into a single 2D NumPy array (909, 768)
embeddings_matrix = np.array(df_final['bert_embeddings'].tolist())

# Final sanity check: ensure no NaNs survived
if np.isnan(embeddings_matrix).any():
    print("⚠️ Found NaNs in matrix. Replacing with 0.0...")
    embeddings_matrix = np.nan_to_num(embeddings_matrix)

print(f"Matrix shape ready for UMAP: {embeddings_matrix.shape}")
```

```python
import umap
import hdbscan
from sklearn.metrics import silhouette_score

# 1. Compress 768D BERT to 5D UMAP
reducer = umap.UMAP(
    n_neighbors=5,       # CRITICAL: Lowering this forces BERT to find tiny, local groups
    n_components=5,      # Keep 5D to allow for clear density separation
    metric='cosine',
    min_dist=0.0,        # Ensure points overlap as little as possible
    random_state=42
)
u_embeddings = reducer.fit_transform(embeddings_matrix)

# 2. Run HDBSCAN on the COMPRESSED data
clusterer = hdbscan.HDBSCAN(
    min_cluster_size=5,  # Smallest possible valid "topic"
    min_samples=1,       # Don't ignore "borderline" points
    cluster_selection_method='leaf', # Pick the most specific sub-clusters
    metric='euclidean'
)
new_labels = clusterer.fit_predict(u_embeddings)

# 3. Re-calculate Silhouette (excluding noise -1 for a fair test)
mask = new_labels != -1
if mask.any():
    score = silhouette_score(u_embeddings[mask], new_labels[mask])
    print(f"New Silhouette Score: {score:.4f}")
    print(pd.Series(new_labels).value_counts())
```

```python
cluster_stats_list = []

for week in sorted(filtered_df['week'].unique()):
    weekly_df = filtered_df[filtered_df['week'] == week]

    for cluster_id in sorted(weekly_df['cluster_id'].unique()):
        # Exclude noise points (cluster_id == -1)
        if cluster_id != -1:
            cluster_df = weekly_df[weekly_df['cluster_id'] == cluster_id]

            number_of_articles = len(cluster_df)

            # Ensure there are articles in the cluster before calculating mean embedding
            if number_of_articles > 0:
                # Convert list of embeddings to a NumPy array for mean calculation
                embeddings_array = np.array(cluster_df['bert_embeddings'].tolist())
                mean_embedding = embeddings_array.mean(axis=0)

                cluster_stats_list.append({
                    "week": week,
                    "cluster_id": cluster_id,
                    "count": number_of_articles,
                    "centroid": mean_embedding.tolist() # Store as list for DataFrame compatibility
                })

# Convert the list of dictionaries to a DataFrame for easier analysis
cluster_stats_df = pd.DataFrame(cluster_stats_list)

print("Cluster statistics calculated and stored in 'cluster_stats_df'.")
display(cluster_stats_df.head())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

cluster_trends_data = []

for cluster_id in sorted(filtered_df['cluster_id'].unique()):
    if cluster_id == -1: # Skip noise points
        continue

    cluster_data = cluster_stats_df[cluster_stats_df['cluster_id'] == cluster_id].sort_values(by='week')

    # Use .apply() to convert each Period object in the Series to a Timestamp
    cluster_data['week_timestamp'] = cluster_data['week'].apply(lambda x: x.to_timestamp())

    if len(cluster_data) < 2:
        continue

    first_week_timestamp = cluster_data['week_timestamp'].iloc[0]

    prev_count = None
    prev_centroid = None
    prev_growth = None

    for index, row in cluster_data.iterrows():
        current_week = row['week']
        current_week_timestamp = row['week_timestamp']
        current_count = row['count']
        current_centroid = np.array(row['centroid']).reshape(1, -1)

        age_days = (current_week_timestamp - first_week_timestamp).days

        growth = 0.0
        drift = 0.0
        acceleration = 0.0

        if prev_count is not None:
            # Calculate growth
            if prev_count != 0:
                growth = (current_count - prev_count) / prev_count
            else:
                growth = np.inf # Handle division by zero

            # Calculate drift
            drift = 1 - cosine_similarity(current_centroid, prev_centroid)[0][0]

            # Calculate acceleration
            if prev_growth is not None:
                acceleration = growth - prev_growth

            cluster_trends_data.append({
                "week": current_week,
                "cluster_id": cluster_id,
                "age_days": age_days,
                "growth": growth,
                "drift": drift,
                "acceleration": acceleration
            })

        prev_count = current_count
        prev_centroid = current_centroid
        prev_growth = growth

cluster_trends_df = pd.DataFrame(cluster_trends_data)

print("Cluster trends calculated successfully.")
display(cluster_trends_df.head())
```

```python
def classify_stage(age_days, growth, acceleration, drift):

    if growth < -0.2:
        return "declining" # decay detection

    if age_days <= 30:
        if growth > 0.3 or drift > 0.2:
            return "emerging"

    elif 30 < age_days <= 90:
        if growth > 0.1:
            return "trending"

    elif age_days >= 180:
        if acceleration < 0.05:
            return "established"

    return "stable"
```

```python
cluster_trends_df['stage'] = cluster_trends_df.apply(lambda row: classify_stage(
    row['age_days'],
    row['growth'],
    row['acceleration'],
    row['drift']
), axis=1)

print("Cluster stages classified successfully.")
display(cluster_trends_df.head())
```

```python
cluster_trends_df['risk_score'] = (
    0.4 * cluster_trends_df['growth'] +
    0.3 * cluster_trends_df['acceleration'] +
    0.3 * cluster_trends_df['drift']
)

print("Risk scores calculated and added to 'cluster_trends_df'.")
display(cluster_trends_df.head())
```
